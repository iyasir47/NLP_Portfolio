**TASK:1**

```python
# insert your code here


import nltk
import torch
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')


#Load the model
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-mpnet-base-v2")


from google.colab import files
uploaded = files.upload()
file_path = "Alice.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
print("File read successfully!")


sentences = nltk.sent_tokenize(text)  #Split text into sentences


sentence_embeddings = model.encode(sentences, convert_to_tensor=True)  #Compute embeddings for all sentences

def search_similar_sentences(query, top_n=5):
    """
    Finds and prints the top `top_n` most similar sentences from the book.
    """

    query_embedding = model.encode([query], convert_to_tensor=True) #Compute embedding for the query sentence


    similarities = cosine_similarity(query_embedding.cpu(), sentence_embeddings.cpu())[0]  #Compute cosine similarity


    top_indices = similarities.argsort()[-top_n:][::-1] #Get top `top_n` most similar sentences (excluding the query itself if present)

    print("\nTop 5 most similar sentences:\n")
    for idx in top_indices:
        print(f"{sentences[idx]}\n")

#Examples query
query_sentence = "The rabbit was in a hurry."
search_similar_sentences(query_sentence)
```

```python
#insert your code here
''' Define four different search prompts '''
search_prompts = [
    "A strange creature is talking to Alice.",
    "Alice is lost in a mysterious place.",
    "The Queen is angry and gives an order.",
    "Alice drinks something and changes size."
]

''' Run search function for each prompt '''
print("\n===== English Search Results =====\n")
for prompt in search_prompts:
    print(f"\nüîç Query: {prompt}")
    search_similar_sentences(prompt)
    print("=" * 80)

 ''' Define translated versions of the search prompts (e.g., Spanish) '''
translated_prompts = [
    "Una criatura extra√±a est√° hablando con Alicia.",    #Strange creature talking to Alice
    "Alicia est√° perdida en un lugar misterioso.",       #Alice is lost in a mysterious place
    "La Reina est√° enojada y da una orden.",             #The Queen is angry and gives an order
    "Alicia bebe algo y cambia de tama√±o."               #Alice drinks something and changes size
]

''' Run search function for each translated prompt '''
print("\n===== Spanish Search Results =====\n")
for prompt in translated_prompts:
    print(f"\nüîç Query: {prompt}")
    search_similar_sentences(prompt)
    print("=" * 80)
```
